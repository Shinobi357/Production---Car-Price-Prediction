# Car Price Prediction API Demo

This notebook demonstrates how to interact with the production Flask API.

The API accepts vehicle information and returns an estimated selling price.

```python
import requests
```

```python
url = "http://127.0.0.1:5000/predict"

payload = {
    "Brand": "Toyota",
    "Year": 2018,
    "Engine Size": 2.5,
    "Fuel Type": "Petrol",
    "Transmission": "Automatic",
    "Mileage": 55000,
    "Condition": "Used",
    "Model": "Camry"
}

response = requests.post(url, json=payload)

response.json()
```

The API successfully returned an estimated vehicle price.

This demonstrates the deployed machine learning model can receive structured JSON input and produce real-time predictions.
